## Turistkonsum

Hele beløpet som besøkende bruker på produkter og tjenester mens de er på reise eller ferie, kalles turistkonsum. Dette omfatter kostnadene for den besøkendes overnatting, måltider, transport, aktiviteter og andre produkter og tjenester som de måtte betale for under besøket.

I sesongene før pandemien påvirket en rekke variabler veksten av turisme og dens innvirkning på Norges overnattingssystem. For det første var det en vekst i antall utenlandske besøkende trukket av Norges mangfoldige landskap og rike kultur. For det andre var det en økning i bruken av nettbaserte bookingtjenester som Booking.com og Airbnb, noe som gjorde det enklere for reisende å reservere og betale for overnatting i Norge. Som et svar har markedet sett både økt etterspørsel og konkurranseevne.

Prisene på overnatting og andre tilhørende tjenester i Norge økte som følge av landets økende etterspørsel etter begge. Hoteller og andre overnattingssteder måtte konkurrere med hverandre om pengene til reisende, noe som økte konkurransen i virksomheten.

Pandemien hadde imidlertid en skadelig innvirkning på overnattingssektoren i 2020 og 2021. Etterspørselen etter overnatting og reiselivstjenester gikk ned som følge av reiserestriksjoner og nedgang i antall besøkende. Som et resultat, for å trekke gjester og sikre lønnsomhet, måtte hoteller og andre overnattingssteder senke prisene og gi ut tilbud. Dette kan karakteriseres som en nedgang i etterspørselen etter overnattingstjenester og en justering av hotelltilbudene.


```python
postUrl = "https://data.ssb.no/api/v0/no/table/10638/"

apiQuery = {
  "query": [
    {
      "code": "KonsumInnd",
      "selection": {
        "filter": "item",
        "values": [
          "UTLK"
        ]
      }
    },
    {
      "code": "Reiseliv",
      "selection": {
        "filter": "item",
        "values": [
          "01.01"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "Priser"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "item",
        "values": [
          "2017",
          "2018",
          "2019",
          "2020",
          "2021"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
import matplotlib.pyplot as plt

# Data
tid = [2017, 2018, 2019, 2020, 2021]
priser = [1447, 1429, 1431, 1369, 1417]

# Plot
plt.plot(tid, priser)
plt.title('Turistkonsum i Norge')
plt.xlabel('År')
plt.ylabel('Løpende priser (mill. kr)')
plt.grid(True)
plt.xlim(2017, 2021)  # Justerer x-aksen
plt.ylim(min(priser)-50, max(priser)+50)  # Justerer y-aksen
plt.xticks(tid, [str(t) for t in tid])
plt.show()
```

Nedgangen i etterspørselen etter overnattingstjenester som følge av reiserestriksjoner og økende smittefrykt var en av de viktigste faktorene som påvirket overnattingsbransjen. Som et resultat ble det færre bestillinger og besøkende, noe som igjen resulterte i en nedgang i industriens omsetning.

Resultatet av dette var en nedgang i prisene ettersom hotellene måtte senke prisene for å lokke gjester og bevare lønnsomheten. Det var også mer konkurranse fra alternative typer overnatting, som Airbnb, som gir rimeligere og mer tilpasningsdyktige overnattingsmuligheter.
Ettersom det var et merkbart fall i bestillinger når prisene steg, karakteriseres etterspørselen etter losjitjenester som elastisk. Nedgangen i pris kan også sees på som en reaksjon på variasjoner i etterspørselen.

Hotellene kan karakteriseres som prissensitive fra tilbudssiden siden de må endre prisene for å tilfredsstille etterspørselen. Et annet eksempel på en markedsforstyrrelse som påvirker tilbudssiden av markedet er konkurranse fra alternative overnattingsmuligheter.

## Kapitel 3

## Hva er formålet med å innføre en turistskatt på overnattinger?

formålet med å innføre en turistskatt på overnattinger kan være å øke statens inntekter eller å regulere turiststrømmen. innføringen av en turistskatt endrer prisstrukturen for turistnæringen. Dette kan påvirke både tilbud og etterspørsel. På tilbudssiden kan det hende at noen hoteller og overnattingssteder vil redusere antall tilgjengelige rom for å unngå å måtte betale en høyere turistskatt. Dette kan føre til en reduksjon i tilbudet av overnattingssteder, og dermed øke prisene for de gjenværende rommene. På etterspørselssiden kan det hende at noen turister vil velge å ikke reise til et område på grunn av den høyere turistskatten. Dette kan føre til en reduksjon i etterspørselen etter overnatting, og dermed redusere prisene på overnatting for de som fortsatt ønsker å besøke området. På den andre siden kan innføringen av en turistskatt også øke statens inntekter. Dette kan gi staten muligheten til å investere i infrastruktur og tjenester som kan øke turismen ytterligere.

## Hvordan fungerer markedet for overnattinger uten en skatt?
Priser i et marked uten avgift på losji vil avgjøres av tilbud og etterspørsel. Hoteller og andre overnattingssteder vil da konkurrere om forbrukerne basert på pris, kvalitet og beliggenhet.

Skatter på overnattingstjenester vil øke kostnadene for overnatting siden hoteller og andre overnattingsfasiliteter vil pådra seg flere utgifter som de vil prøve å kompensere ved å øke prisene. Dette vil ha innvirkning på etterspørselen siden økte kostnader vil hindre turister fra å kjøpe. Dette kan også tvinge overnattingssteder som hoteller til å endre seg, kanskje ved å redusere antall tilgjengelige rom eller innføre spesialtilbud for å trekke inn gjester.

Uten skatt vil tilbud og etterspørsel av overnattingstjenester avgjøre hvor mye de koster. En avgift vil øke prisnivået og ha innvirkning på både tilbud og etterspørsel på overnattingstjenester.
